```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
credit_train = pd.read_csv('C:\\Users\\monte\\Downloads\\credit_scoring\\credit_train.csv\\credit_train.csv')
credit_train
```

```python
credit_train = credit_train.drop_duplicates(['Loan ID','Customer ID'])
```

```python
credit_train.info()
```

```python
credit_train = credit_train.rename(columns = {'Loan ID':'Loan_ID','Customer ID':'Customer_ID','Loan Status':'Loan_Status',
                                              'Current Loan Amount':'Current_Loan_Amount','Credit Score':'Credit_Score',
                                              'Annual Income':'Annual_Income','Years in current job':'Years_in_current_job',
                                              'Home Ownership':'Home_Ownership','Monthly Debt':'Monthly_Debt',
                                              'Years of Credit History':'Years_of_Credit_History',
                                              'Months since last delinquent':'Months_since_last_delinquent',
                                              'Number of Open Accounts':'Number_of_Open_Accounts',
                                              'Number of Credit Problems':'Number_of_Credit_Problems',
                                              'Current Credit Balance':'Current_Credit_Balance',
                                              'Maximum Open Credit':'Maximum_Open_Credit','Tax Liens':'Tax_Liens'})
```

```python
credit_train.loc[credit_train.Current_Loan_Amount > 1000000, 'Current_Loan_Amount'] = credit_train.Current_Loan_Amount.median()

credit_train.loc[credit_train.Credit_Score > 3000, 'Credit_Score'] = credit_train.Credit_Score.median()

credit_train.loc[credit_train.Annual_Income > 400000, 'Annual_Income'] = credit_train.Annual_Income.median()
```

```python
credit_train.loc[credit_train.Current_Credit_Balance > 10000000, 'Current_Credit_Balance'] = 10000000

credit_train.loc[credit_train.Maximum_Open_Credit > 150000000, 'Maximum_Open_Credit'] = credit_train.Maximum_Open_Credit.median()
```

```python
credit_train.isnull().sum()
```

```python
credit_train.Annual_Income = credit_train.Annual_Income.fillna(credit_train.Annual_Income.median())
credit_train.Credit_Score = credit_train.Credit_Score.fillna(credit_train.Credit_Score.median())
credit_train.Maximum_Open_Credit = credit_train.Maximum_Open_Credit.fillna(credit_train.Maximum_Open_Credit.median())
```

```python
credit_train.Years_in_current_job = credit_train.Years_in_current_job.fillna(-100)
credit_train.Months_since_last_delinquent = credit_train.Months_since_last_delinquent.fillna(-100)
credit_train.Bankruptcies = credit_train.Bankruptcies.fillna(-100)
credit_train.Tax_Liens = credit_train.Tax_Liens.fillna(-100)
```

```python
X_train = credit_train.drop(['Loan_ID', 'Customer_ID','Loan_Status'], axis = 1)
y_train = credit_train.Loan_Status
```

```python
X_train = pd.get_dummies(X_train)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
clf = RandomForestClassifier(random_state=123)
```

```python
params = {'criterion': ['gini','entropy'],
          'min_samples_split': [10,50,100],
          'max_leaf_nodes' : [128,256,512],
          'n_estimators': [250,500,1000]}
```

```python
searcher = RandomizedSearchCV(clf, params, cv = 5, n_iter = 5, scoring = 'accuracy')
```

```python
searcher.fit(X_train,y_train)
```

```python
searcher.best_params_
```

```python
clf1 = searcher.best_estimator_
```

```python
clf1.fit(X_train,y_train)
```

```python
clf1.score(X_train,y_train)
```
